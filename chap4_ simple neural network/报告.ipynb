**实验报告**

### 1. 函数定义
f(x) = sin(2 \pi x) + 0.5 x^2 

### 2. 数据
- **训练集**：在区间 \([0, 1]\) 上均匀采样 500 个样本点。
- **测试集**：在区间 \([0, 1]\) 上均匀采样 100 个样本点。


### 3. 模型描述
网络结构：
- 输入层：1个神经元，接收标量输入。
- 隐藏层1：64个神经元，激活函数`ReLU`。
- 隐藏层2：64个神经元，激活函数`ReLU`。
- 输出层：1个神经元，输出标量预测值。

此外，
- **损失函数** 采用 `MSELoss`（均方误差）。
- **优化器** 采用 `Adam`（学习率 0.01）。


### 4. 实验结果
可视化结果见.\pytorch_func_sim.ipynb

模型在整体上较好地拟合了目标函数，尤其在非线性变化剧烈的区间（如正弦波峰、波谷处）拟合效果较为优秀。然而，在靠近边缘部分存在一定误差，这可能是由于训练样本密度不足或模型容量有限所致。
